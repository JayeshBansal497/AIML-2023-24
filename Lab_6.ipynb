K-Nearest Neighbor Classification Using Classified Manufacturing Dataset

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_table("/content/sample_data/Classified_Data.txt",sep=',',index_col=0)
dataset.head()
```

```python
dataset.info(verbose=True)
```

```python
dataset.describe(percentiles=[.25,.5,.75,.9])
```

```python
dataset.columns
```

```python
list1=list(dataset.columns)
list1[0:len(list1)-1]
for i in range(len(list1)-1):
  sns.boxplot(x='TARGET CLASS',y=list1[i],data=dataset,color='cyan')
  plt.figure()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(dataset.drop('TARGET CLASS',axis=1))
scaled_features=scaler.transform(dataset.drop('TARGET CLASS',axis=1))
```

```python
data_feat=pd.DataFrame(scaled_features,columns=dataset.columns[:-1])
data_feat.head()
```

Part 2 - Model Training

```python
from sklearn.model_selection import train_test_split
x=data_feat
y=dataset['TARGET CLASS']
x_train,x_test,y_train,y_test=train_test_split(scaled_features,dataset['TARGET CLASS'],test_size=0.30,random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train,y_train)
```

Part 3 - Evaluatation

```python
pred=knn.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
conf=confusion_matrix(y_test,pred)
print(conf)
```

```python
print(classification_report(y_test,pred))
```

```python
print('MISCLASSIFICATION ERROR RATE',round(np.mean(pred!=y_test),3))
```

```python
error_rate=[]
for i in range(1,60):
  knn=KNeighborsClassifier(n_neighbors=i)
  knn.fit(x_train,y_train)
  pred_i=knn.predict(x_test)
  error_rate.append(np.mean(pred_i!=y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,60),error_rate,color='blue',linestyle='dotted',marker='o',markerfacecolor='red',markersize=8)
plt.title('Error rate v/s k-value',fontsize=20)
plt.xlabel('k',fontsize=15)
plt.ylabel('Error(Misclassification) Rate',fontsize=15)
```
